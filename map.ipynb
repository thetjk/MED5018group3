```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pyecharts.charts import Line,Bar,Map,Timeline
from pyecharts.commons.utils import JsCode
import pyecharts.options as opts
from pyecharts.globals import ThemeType

df1 = pd.read_csv("C:\\Users\\Einstein Lee Cooper\\Desktop\\python\\aids\\aids\\tb-patients-tested-positive-for-hiv.csv")
df2 = pd.read_csv("C:\\Users\\Einstein Lee Cooper\\Desktop\\python\\aids\\aids\\tb-related-deaths-hiv.csv")

print(df1.columns)
print(df2.columns)

year_1 = [i for i in range(2004,2011)]
year_2 = [i for i in range(2000,2017)]

timeline = Timeline(init_opts=opts.InitOpts(theme=ThemeType.DARK, width='1500px', height='850px'))
for year in year_1:
    temp = df1[df1['Year']==year]
    National_data = temp['Entity'].values.tolist()

    data = temp['TB patients tested positive for HIV - Total'].values.tolist()

    # 使用zip压缩为元组形式
    qz_data = [tuple(z) for z in zip(National_data, data)]

    # 设置副标题
    subtitle = "%d："%year  + "TB patients tested positive for HIV - Total"

    world_map_data = (Map(
            # 设置地图宽度和高度
            init_opts={"width": "1280px", "height": "800px"})
            .add('HIV',qz_data,maptype="world",is_selected=True,is_map_symbol_show=False)
            # 系列配置
            .set_series_opts(
            # 不显示经纬度，设置颜色，字体大小
            label_opts={'is_show': 'False','color': '#fff', 'font_size': '18', 'position': 'left'})
            # 整体的配置
            .set_global_opts(
            # 设置左下方的颜色标注
            visualmap_opts=opts.VisualMapOpts(min_=0, max_=int(8000)),
            # 设置左上角标题和副标题
            title_opts=opts.TitleOpts(
                title="TB patients tested positive for HIV - Total", subtitle=subtitle, pos_left="50", pos_top="5%",
                title_textstyle_opts=opts.TextStyleOpts(font_size=30),
                subtitle_textstyle_opts=opts.TextStyleOpts(font_size=18, color='#3')),
            # 图例设置
            legend_opts=opts.LegendOpts(selected_mode='single', pos_top="50", pos_bottom="5%",textstyle_opts=opts.TextStyleOpts(font_size=18))))
    timeline.add(chart=world_map_data, time_point='%d'%year)

    ### 设置轮播时间和轮播间隔
    timeline.add_schema(is_auto_play=True, play_interval=1000)

timeline.render("tb-patients-tested-positive-for-hiv.html")

timeline = Timeline(init_opts=opts.InitOpts(theme=ThemeType.DARK, width='1500px', height='850px'))
for year in year_2:

    temp = df2[df2['Year']==year]
    Nationa2_data = temp['Entity'].values.tolist()

    data = temp['Estimated TB-related deaths among people living with HIV - estimate'].values.tolist()

    # 使用zip压缩为元组形式
    qz_data = [tuple(z) for z in zip(Nationa2_data, data)]

    # 副标题
    subtitle = "%d："%year  + "Estimated TB-related deaths among people living with HIV - estimate"

    world_map_data = (Map(
            # 设置宽度，高度
            init_opts={"width": "1280px", "height": "800px"})
            #在选中目标时加上前缀HIV          
            .add('HIV',qz_data,maptype="world",is_selected=True,is_map_symbol_show=False)
            # 系列配置
            .set_series_opts(
            # 不显示经纬度，设置颜色，字体大小
            label_opts={'is_show': 'False','color': '#fff', 'font_size': '18', 'position': 'left'})
            # 整体的配置
            .set_global_opts(
            # 设置左下方的颜色标注
            visualmap_opts=opts.VisualMapOpts(min_=0, max_=int(6000)),
            # 设置左上角标题和副标题
            title_opts=opts.TitleOpts(title="Estimated TB-related deaths among people living with HIV - estimate", subtitle=subtitle, pos_left="50", pos_top="5%",
            title_textstyle_opts=opts.TextStyleOpts(font_size=30),subtitle_textstyle_opts=opts.TextStyleOpts(font_size=18, color='#3')),
            # 图例设置
            legend_opts=opts.LegendOpts(selected_mode='single', pos_top="50", pos_bottom="5%",textstyle_opts=opts.TextStyleOpts(font_size=18))))
    timeline.add(chart=world_map_data, time_point='%d'%year)

    ### 设置轮播时间和轮播间隔
    timeline.add_schema(is_auto_play=True, play_interval=1000)

timeline.render("tb-related-deaths-hiv.html")
```
