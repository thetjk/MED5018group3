```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the csv file
df = pd.read_csv('deaths-and-new-cases-of-hiv.csv')
# country subset
df_cn = df[df['Code'] == 'CHN']
df_us = df[df['Code'] == 'USA']
```

```python
#extract data
year = df_cn['Year'].tolist()
cn_cases = df_cn['Incidence - HIV/AIDS - Sex: Both - Age: All Ages (Number)'].tolist()
us_cases = df_us['Incidence - HIV/AIDS - Sex: Both - Age: All Ages (Number)'].tolist()
cn_deaths = df_cn['Deaths - HIV/AIDS - Sex: Both - Age: All Ages (Number)'].tolist()
us_deaths = df_us['Deaths - HIV/AIDS - Sex: Both - Age: All Ages (Number)'].tolist()
```

```python
fig,ax = plt.subplots()
ax.plot(year,cn_cases,color='yellow',marker='*',linestyle='-',label='CN')
ax.plot(year,us_cases,color='fuchsia',marker='+',linestyle='--',label = 'US')
month_label = year
ax.set_xticks(year)
ax.set_xticklabels(month_label,rotation=45)
ax.set_xlabel('years')
ax.set_ylabel('new cases')
ax.set_title('HIV Yearly New Cases')
ax.grid(axis='both')
ax.legend(loc = 3, title = 'Cases')

ax2 = ax.twinx()
ax2.plot(year,cn_deaths,color='green',marker='*',linestyle='-',label='CN')
ax2.plot(year,us_deaths,color='tomato',marker='+',linestyle='--',label = 'US')
ax2.set_ylabel('new deaths')
ax2.legend(loc = 4,title = 'Deaths')

#save the plot
fig.savefig('HIV New Cases.png',dpi = 400,bbox_inches = 'tight',facecolor='white')
```
